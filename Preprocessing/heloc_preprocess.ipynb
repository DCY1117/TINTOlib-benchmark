```python
import pandas as pd
```

```python
df = pd.read_csv('../Datasets/heloc.csv')
```

```python
# Check the first few rows of the DataFrame
print("First few rows:")
print(df.head())
```

```python
# Get information about the DataFrame
print("\nDataFrame information:")
print(df.info())
```

```python
# Get summary statistics of numerical columns
print("\nSummary statistics:")
print(df.describe())
```

```python
# Check the distribution of the target variable (assuming it's named 'target')
print("\nDistribution of the target variable:")
print(df['RiskPerformance'].value_counts())
```

```python
# Assuming 'target' is the label column
label_column = 'RiskPerformance'

# Identify rows where all values are -9 (excluding the label column)
rows_to_eliminate = (df.drop(columns=[label_column]) == -9).all(axis=1)

# Print the number of rows to be eliminated
print(f"\nNumber of rows to be eliminated: {rows_to_eliminate.sum()}")

if rows_to_eliminate.sum() > 0:
    # Print the rows to be eliminated
    print("Rows to be eliminated:")
    print(df.loc[rows_to_eliminate])
    
    # Eliminate the rows
    df = df.loc[~rows_to_eliminate]
    print("Rows eliminated.")
else:
    print("No rows found where all values are -9 (excluding the label column).")
```

```python
# Save processed DataFrame as CSV
df.to_csv('../Datasets/processed_heloc.csv', index=False)
```
